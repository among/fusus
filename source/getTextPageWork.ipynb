## LAM-ALEF(wasla) does not come out right

Here is an example

![lamalefwasla](lamalefwasla.png)

See line 10 below.

```python
testPages = "34"
char = "\ue809"
Lw.drawPages(testPages)
```

```python
Lw.showLines(testPages, search=char, orig=True)
```

If we render the string, we do not get the right ligature.

Whereas when we replace the alef-wasla with an ordinary alef, the ligature comes out right:

```python
alefWasla = "\u0671"
alefNorml = "\u0627"

Lw.showString(f"\u0644\u0650\u0644\u0652{alefWasla}\u0633\u0652\ufe98")
Lw.showString(f"\u0644\u0650\u0644\u0652{alefNorml}\u0633\u0652\ufe98")
```
