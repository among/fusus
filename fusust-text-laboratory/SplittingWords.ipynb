```python
def InsertRow(after, df, columns, values):
    df1 = df.copy()[0:after+1]
    df2 = df.copy()[after+1:]
    df1.loc[after+1]=df1.loc[after]
    df1.loc[after+1,columns]=values
    result = pd.concat([df1,df2])
    result.index = [*range(result.shape[0])]
    return result
```

```python
def SplitRow(df, row, splitWord, splitShort, columns1, values1, columns2, values2):
    df1 = df.copy()[0:row+1]
    df2 = df.copy()[row+1:]
    df1.loc[row+1]=df1.loc[row]
    df1.loc[row+1,'word'] = df1.loc[row,'word'][splitWord:]
    df1.loc[row,'word'] = df1.loc[row,'word'][:splitWord]
    df1.loc[row+1,'short'] = df1.loc[row,'short'][splitShort:]
    df1.loc[row,'short'] = df1.loc[row,'short'][:splitShort]
    df1.loc[row,columns1]=values1
    df1.loc[row+1,columns2]=values2
    result = pd.concat([df1,df2])
    result.index = [*range(result.shape[0])]
    return result
```
