# Page 399, line 4

When calling up the image as drawn by rendering the PDF, we specify a vertical region and pass it in the
`clip` parameter: top and bottom.
It is a bit of trial and error to find the right `y`-coordinates.

```python
(pg, ln) = (399, 4)
Lw.drawPages(pg, clip=(60, 190))
```

```python
Lw.getPages(pg)
Lw.htmlPages(pg, line=ln)
```
