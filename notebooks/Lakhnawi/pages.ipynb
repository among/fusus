```python
Lw = Lakhnawi()
Lw.setStyle()
```

# Title page

We can draw a page from the PDF as an image.
Note that the PDF is a textual PDF, it does not contain images.
Drawing a page means: *rendering* the page as a raster image with a fixed resolution.

It is the same what Acrobat Reader or Apple Preview does when it displays a PDF.

```python
testPages = "1"
Lw.drawPages(testPages)
```

Let's fetch the data of this page from the PDF:

```python
Lw.getPages(1)
```

And show it as an HTML page based on the extracted text:

```python
Lw.htmlPages(1)
```

That does not look to bad!

Dividing the characters of the page into lines is not completely trivial.
We follow the process in its tracks:

```python
Lw.showLineHeights(1)
```

The raw heights show how many characters appear at which height (in pixels).

The line heights are obtained by clustering all these heights in reasonable clusters.

Look at line 8: the characters in it have quite a range of different heights.
Especially the last character of the line is an outlier, that is on the verge of being classified as a separate line.

We can write the textual data to a file on disk in TSV format: each word occupies a line, and its data is in tab-separated fields.

```python
Lw.tsvPages(1)
```

# All pages

We can do this also for all pages, and export the results to disk.

```python
testPages = None
Lw.getPages(testPages)
```

We write the complete result to HTML files on disk, in several flavours:

* one file per page
* single file
* single file, with table of contents

```python
Lw.htmlPages(testPages, export=True, singleFile=False)
Lw.htmlPages(testPages, export=True, singleFile=True)
Lw.htmlPages(testPages, export=True, singleFile=True, toc=True)
```

We write the complete TSV data to disk:

```python
Lw.tsvPages(testPages)
```
